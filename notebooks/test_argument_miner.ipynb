```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure, from_networkx
from bokeh.models import Range1d, Circle, MultiLine
from networkx.drawing.nx_agraph import graphviz_layout
import networkx as nx

plot = figure(tools='pan, wheel_zoom, save, reset', active_scroll='wheel_zoom', x_range=Range1d(-10.1, 10.1), y_range=Range1d(-10.1, 10.1), width=1100, height=600)


pos = graphviz_layout(am.network, prog='sfdp')
pos = nx.rescale_layout_dict(pos, scale=10)

network_graph = from_networkx(am.network, pos, scale=10, center=(0, 0))

network_graph.node_renderer.glyph = Circle(size=10, fill_color='color')
network_graph.edge_renderer.glyph = MultiLine(line_alpha=0.1, line_width=1)

plot.renderers.append(network_graph)
output_notebook()
show(plot)
```
