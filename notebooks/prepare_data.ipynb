```python
import pandas as pd
```

```python
dpath = "./AMAZON_FASHION_5.json"
df = pd.read_json(dpath, lines=True)
df['reviewText'] = df['reviewText'].fillna('')
```

```python
df_review = df[['asin', 'reviewText']]
df_review = df_review.iloc[df_review.astype(str).drop_duplicates().index]
df_review.groupby(by='asin')['reviewText'].count().reset_index(name='count').sort_values(by='count', ascending=False)
```

```python
# pick the product with the highest number of unique reviews
asin_selected = 'B009MA34NY'
df_review_selected = df_review[df_review['asin'] == asin_selected].reset_index(drop=True)
df_review_selected
```

```python
df_review_selected.to_json('./sample_data.json', lines=True, orient='records')
```
